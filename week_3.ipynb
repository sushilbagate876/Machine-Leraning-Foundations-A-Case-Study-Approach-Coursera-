```
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```
def awesome_count(x):
  count = 0
  for word in x:
    if word == 'awesome':
      count += 1
  return count
```

```
products['awesome'] = products['word_count'].apply(awesome_count)
```

```
products['awesome'].sum()
```
